```python
#Now, let's start the data cleaning !!
```

```python
#Since we know many age data are missing, we have to do something about it.
#It can be either by taking a mean of all the ages, or we can take the average age separating by class!
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=train,palette='winter')
```

```python
#Here we can see that, people belonging to class 1 are older on average and people belong to class 3 are the youngest
#Now we are going to impute the age
#From the boxplot, mean age of class 1 is 37, class 2 is 29 and class 3 is about 24
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
#now we check if there are any missing data in age
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#Now, we are going to drop the cabin column as it's not that influential of a column
```

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.info()
```

```python
train.head(30)
```

```python
#Sex and Embarked are categorical columns . We need to convert them into strings.
#Get dummies method needs to be used
```

```python
sex= pd.get_dummies(train['Sex'],drop_first=True)
embark = pd.get_dummies(train['Embarked'],drop_first=True)

#by doing this, we got their dummy values
```

```python
#Now we drop the categorical columns or string columns
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
#concatinating our new sex and embark column with numeric values
train = pd.concat([train,sex,embark],axis=1)
```

```python
train.head(30)
```

```python
train.dropna(inplace=True)
```

```python
train.info()
```

```python
#Now we have all the data 
```

```python
#ML tranining
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(train.drop('Survived',axis=1),
                                                train['Survived'],test_size=0.30, random_state=101)
```

```python
#Prediction
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
#Evaluation
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
#Now we use decision tree to see if we can get better result
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
dtree_predict= dtree.predict(X_test)
print(classification_report(y_test,dtree_predict))
print(confusion_matrix(y_test,dtree_predict))
```

```python
rf=RandomForestClassifier(n_estimators=100)
```

```python
rf.fit(X_train,y_train)
rf_predict= rf.predict(X_test)
print(classification_report(y_test,rf_predict))
print(confusion_matrix(y_test,rf_predict))
```
